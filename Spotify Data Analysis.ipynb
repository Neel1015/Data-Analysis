# Make the Release Date Column as the Index Column

```python
# We will perform this action with the help of the set_index function

df_tracks.set_index("release_date",inplace=True)
df_tracks.index = pd.to_datetime(df_tracks.index)
df_tracks.head()
```

# Convert the Duration of the Songs From Milliseconds to Seconds

```python
# We will convert the duration of the songs from milliseconds to seconds and verify it by printing the headings of the dataset to check whether the duration is converted into seconds

df_tracks["duration"]= df_tracks["duration_ms"].apply(lambda x: round(x/1000))
df_tracks.drop("duration_ms", inplace=True, axis=1)
```

```python
df_tracks.duration.head()
```

# Correlation Map

```python
# First we will drop three unwanted keys, mode and explicit columns, and apply the pearson correlation method
# We will set the figure size for the correlation map to (14,6). We will use the heatmap() function to create our correlation map, plus we will set the annotation = True that will write the data value in each cell. 
# We will set fmt=" .1g"; this is string formatting quotes used when adding annotations. 
# Here cmap stands for the color map. 
# You can google sns cmap and choose any color from the documentation if you wish
```

```python
corr_df = df_tracks.drop(["key","mode","explicit"], axis=1).corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap= sns.heatmap(corr_df,annot=True,fmt=".1g",vmin=-1,vmax=1,center=0,cmap="inferno",linewidths=1,linecolor="Black")
heatmap.set_title("Correlation Heatmap Between Variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

```python
# After running the piece of code, we got our correlation map. On the right side, you can see a scale ranging from -1 to +1 
# Here -1 denotes the variables that have the least or negative correlation, 
# while the values above 0.0 denote the variables with a positive correlation
```

# Create a Regression Plot Between Loudness and Energy. Let’s Plot It  in the Form of a Regression Line

```python
# We will use the regplot() function present in the seaborn library to draw the regression plot

plt.figure(figsize=(10,6))
sns.regplot(data= sample_df,y="loudness",x="energy",color="c").set(title="Loudness vs Energy Correlation")
```

```python
# The result is plotted. There is a very high positive correlation between loudness and energy. 
# You can also see that all the data points or the songs are in one direction.
# If the energy increases, the loudness of the song increases and similarly,
# if the song's loudness decreases, the energy of the track also decreases
```

#Similarly, we can plot another regression plot between popularity and acousticness

# Create a Regression Plot Between Popularity and Acousticness in the Form of a Regression Line

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y="popularity",x="acousticness",color="b").set(title="Popularity vs Acousticness Correlation")
```

```python
# Here, we can see the blue color regression line is in downward direction, 
# which denotes if the acousticness of the song increases, the popularity decreases and similarly, 
# if the popularity increases, the acousticness decreases
```

```python
df_tracks['dates'] = df_tracks.index.get_level_values('release_date')
df_tracks.dates= pd.to_datetime(df_tracks.dates)
years= df_tracks.dates.dt.year
```

```python
# Now, we will use the seaborn library and the linepolt function
```

# Plot a Line Graph to Show the Duration of the Songs for Each Year

```python
total_dr= df_tracks.duration
sns.set_style(style="whitegrid")
fig_dims= (10,5)
fig,ax = plt.subplots(figsize=fig_dims)
fig= sns.lineplot(x=years,y=total_dr,ax=ax).set(title="Year vs Duration")
plt.xticks(rotation=60)
```

```python
# We got the line plot. On the X-axis, 
# we have the years and on the Y-axis, 
# we have the duration. 
# Here, we can see the songs from the 1920s to 1960s were of shorter duration. 
# After 1960, the duration of the songs started increasing until 2010. From 2010 onwards, 
# the duration again started declining
```

```python
# Here, we got our dataset.

# Plot Duration of the Songs w.r.t. different Genres using a horizontal barplot.
# Here we will use the barplot function present in the seaborn library
```
