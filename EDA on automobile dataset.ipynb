Missing value Detection and treatment

```python
data.info()
```

There are no blank records. The missing records are represented as "?".

```python
# some of the predictor columns are continuous in nature there might be a chance that 0's in these columns indicate missing data
# lets check
data.describe()
```

None of the continuous column sseem to have missing values represented by 0's.column 'normalized-losses' has missing values.

```python
# Treating missing values at normalized-losses
# replacing '?' to NaN
from numpy import nan
data['normalized-losses'] = data['normalized-losses'].replace('?',np.nan)
```

```python
# Lets check the changes
data.head()
```

```python
# Now lets see isna function
data.isna().sum()
```

Normalized losses has 41 missing data points. we will replace these missing values with the median of 'normalized-losses'.Median is least affected by outliers.

```python
# Imputing missing values with their respective columns median
data.fillna(data.median(),inplace=True)
```

```python
data.median()
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.info()
```

**We notice that datatype of normalized-losses is object but we have to changes to numeric i.e. float**

```python
data['normalized-losses']= pd.to_numeric(data['normalized-losses'],downcast='float')
```

```python
data.info()
```

**Now we have treated the missing values lets move to outlier detection and treatment**

Boxplots are great way of detecting outliers. Once outliers have been detected they can be imputed with the 5th and 95th percentiles.

```python
data.columns
```

```python
# Outlier detection using boxplot
plt.figure(figsize=(20,15))

plt.subplot(3,4,1)
sns.boxplot(data['normalized-losses'])

plt.subplot(3,4,2)
sns.boxplot(data['symboling'])

plt.subplot(3,4,3)
sns.boxplot(data['wheel-base'])

plt.subplot(3,4,4)
sns.boxplot(data['length'])

plt.subplot(3,4,5)
sns.boxplot(data['width'])

plt.subplot(3,4,6)
sns.boxplot(data['curb-weight'])

plt.subplot(3,4,7)
sns.boxplot(data['engine-size'])

plt.subplot(3,4,8)
sns.boxplot(data['city-mpg'])

plt.subplot(3,4,9)
sns.boxplot(data['highway-mpg'])

plt.subplot(3,4,10)
sns.boxplot(data['height'])

plt.subplot(3,4,11)
sns.boxplot(data['compression-ratio'])
```

From the above boxplots we can infer that out of 11 continuous variables 8 of them has outliers. These outliers will be imputed with the 5th and 95th percentiles.

```python
data.columns
```

```python
data['normalized-losses'] = data['normalized-losses'].clip(lower=data['normalized-losses'].quantile(0.05),upper=data['normalized-losses'].quantile(0.95))
```

```python
data['wheel-base'] = data['wheel-base'].clip(lower=data['wheel-base'].quantile(0.05),upper=data['wheel-base'].quantile(0.89))
data['normalized-losses'] = data['length'].clip(lower=data['length'].quantile(0.05),upper=data['length'].quantile(0.95))
data['width'] = data['width'].clip(lower=data['width'].quantile(0.05),upper=data['width'].quantile(0.95))
data['engine-size'] = data['engine-size'].clip(lower=data['engine-size'].quantile(0.05),upper=data['engine-size'].quantile(0.95))
data['city-mpg'] = data['city-mpg'].clip(lower=data['city-mpg'].quantile(0.05),upper=data['city-mpg'].quantile(0.95))
data['highway-mpg'] = data['highway-mpg'].clip(lower=data['highway-mpg'].quantile(0.05),upper=data['highway-mpg'].quantile(0.95))
data['compression-ratio'] = data['compression-ratio'].clip(lower=data['compression-ratio'].quantile(0.05),upper=data['compression-ratio'].quantile(0.89))
```

```python
# Outlier detection using boxplot
plt.figure(figsize=(20,15))

plt.subplot(3,4,1)
sns.boxplot(data['normalized-losses'])

plt.subplot(3,4,2)
sns.boxplot(data['wheel-base'])

plt.subplot(3,4,3)
sns.boxplot(data['width'])

plt.subplot(3,4,4)
sns.boxplot(data['engine-size'])

plt.subplot(3,4,5)
sns.boxplot(data['city-mpg'])

plt.subplot(3,4,6)
sns.boxplot(data['highway-mpg'])

plt.subplot(3,4,7)
sns.boxplot(data['compression-ratio'])
```

```python
# Label encoding is the process of converting categorical text data into numerical data
```

```python
# Data before label encoding
data['body-style'].head(20)
```

```python
# Label Encoding
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
data['body-style'] = labelencoder.fit_transform(data['body-style'])
```

```python
# data after label encoding
data['body-style'].head(20)
```

After running label encoding code, we can see that the variable body-style has numerical values raging from 0-4.
The problem with label encoding is that it introduces an order between the categories i.e. 0>1>2>3>4 This might confuses the most  model into that convertible is greater than hatchback.
